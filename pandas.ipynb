```python
#analyzing tabular data with pandas 


#downloading a csv file to operate on it 
from urllib.request import urlretrieve
url4 = 'https://gist.githubusercontent.com/aakashns/f6a004fa20c84fec53262f9a8bfee775/raw/f309558b1cf5103424cef58e2ecb8704dcd4d74c/italy-covid-daywise.csv'
urlretrieve(url4, './data/covid.csv')#downloading in the data folder
import pandas as pd 
covid=pd.read_csv('./data/covid.csv')
covid['new_cases']#view of the same data
covid['new_cases'].copy()#copy of the same data
covid
print(covid.loc[243])#to access a speicific row in the data by label
covid.head(5)#return first 5 rows


covid.tail(5)#return last 5 rows
covid.loc[1,'new_tests']#some values are not present represented by nan

#The loc() function is label based data selecting method which means that we have to pass
#the name of the row or column which we want to select. This method includes
#the last element of the range passed in it, unlike iloc(). loc() can accept the boolean data unlike iloc(). 

covid.new_tests.first_valid_index()#gives the first index where there is no nan value
covid.loc[108:113]

 
covid.sample(10)#retrieve random rows from table
covid.loc[243,'new_tests']



covid.new_cases.sum()#provides sum of the total cases 


covid.iloc[[1,2,3,4]]#Selecting rows using integer indices
covid.iloc[0:6,0:2]#Selecting a range of columns and rows simultaneously
pd.set_option("display.max_rows",None)



covid[covid.index.isin(range(1,100))]#return the rows from index 1 to 100



 
```

```python
#querying data

redalert=covid.new_deaths>150#alternate method for this 
covid[redalert]#covid.loc[covid.new_deaths>100]
covid.loc[covid.new_deaths>100]

#difference bw mean of new cases and mean of new tests and showing only deaths column
covid.loc[covid.new_cases < covid.new_tests,'new_deaths']
len(covid.loc[covid.new_cases < covid.new_tests,'new_deaths'])#gives total number of rows that satisfy the condition 
#adding new column to the table 
covid['positive_rate']=covid.new_cases/covid.new_tests
covid.drop(columns='positive_rate',inplace=True)
print(covid)
```

```python
covid.date#its dtype is object so convert to datetime 
covid.date=pd.to_datetime(covid.date)
print(covid.date)#now its dtype is datetime 
```

```python
covid['year']=pd.DatetimeIndex(covid.date).year
covid['month']=pd.DatetimeIndex(covid.date).month
covid['day']=pd.DatetimeIndex(covid.date).day
covid['weekday']=pd.DatetimeIndex(covid.date).weekday
covid
```
