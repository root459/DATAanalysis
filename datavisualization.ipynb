```python
#Plotting Multiple Lines
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years, apples,marker='o',color='red')
plt.plot(years, oranges,marker='x')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
plt.title("crop yield in kanto")
plt.legend(['Apples','Oranges'])
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 3, 4]

# Create a figure and subplots
plt.figure(figsize=(10, 4))

# Create the first subplot (1 row, 2 columns, first subplot)
plt.subplot(1, 2, 1)
plt.plot(x, y1, marker='o', linestyle='-', color='blue', label='y = x^2')
plt.title('Square of Numbers')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Create the second subplot (1 row, 2 columns, second subplot)
plt.subplot(1, 2, 2)
plt.plot(x, y2, marker='s', linestyle='--', color='red', label='y = x')
plt.title('Linear Relationship')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

Historgram
distribution of data by forming bins along the range of the data and then drawing bars to show no. of observations fall in each bin

```python
import numpy as np
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.hist(flower_df.sepal_width,color='red',bins=3);
plt.subplot(2,2,2)
plt.hist(flower_df.sepal_width,color='red',bins=8);
plt.subplot(2,2,3)
plt.hist(flower_df.sepal_width,color='red',bins=[1,2,3,4,5,6]);
plt.subplot(2,2,4)
plt.hist(flower_df.sepal_width,color='green',bins=np.arange(2,5,0.25));
#used for a single column only
#bins take array as input
```

values range from 2.0 to 4.4 and maximum values lie bw 3.0 to 3.5
intervals are called bins here and now learn to control the bins 


MULTIPLE HISTOGRAMS 
make multiple histograms so we have to reduce the opacity 


```python
sentosa_df=flower_df[flower_df.species=='setosa']
versicolor_df=flower_df[flower_df.species=='versicolor']
virginica_df=flower_df[flower_df.species=='virginica']
#note==> if values come like true false then put in the dataframe[dataframe condition]
```

```python
plt.hist(sentosa_df.sepal_width,alpha=0.6,bins=np.arange(2,5,0.25),label='sentosa');
plt.hist(versicolor_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25),label='versicolor');
plt.hist(virginica_df.sepal_width,alpha=0.3,bins=np.arange(2,5,0.25),label='virginica');
plt.legend();#to add legend labels are required 
```

```python
#creating histogram overloapping each other in multiple histogram
data = [sentosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width]
labels = ['Sentosa', 'Versicolor', 'Virginica']
plt.hist(data, bins=10, alpha=0.6, label=labels, stacked=True)
plt.legend()
plt.show()
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
bill_avg_df=tips_df.groupby('day')[['total_bill']].mean()
bill_avg_df
plt.bar(bill_avg_df.index,bill_avg_df.total_bill)
#index is used because after applying group by on which we apply groupby become index in the returned data frame for example 
#in above day become index  day=>index  day=x axis
```

```python
#now to compare the male and female we use hue
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df);
```

Heatmaps

visualize 2d data like matrix or a table using colors 

```python
flight_df=sns.load_dataset("flights")
flight_df
```

```python
plt.plot(flight_df.passengers)#passengers are increasing 
```

```python
flights_df=sns.load_dataset("flights").pivot("month","year","passengers")
#pivot makes it in form of matrix so by heatmap easily visualize
flights_df
```

```python
sns.heatmap(flights_df,fmt='d',annot=True,cmap='Blues')
#fmt parameter stands for "format" and is used to specify the format for annotating the cells of the heatmap. 
#annot shows the number also without fmt it not work
```
